```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
# File location and type
file_location = "D:\GWU\AAA-Fall 2023\8515-Data Engineering for AI\Data Processing Dataset\Malware_dataset.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

## Rank Values with the `.rank()` Method

```python
dataframe = pd.read_csv("/Workspace/Users/g26420236@gwu.edu/Lab5/Python_DataFrame/revenue.csv", index_col = "Date")
dataframe["Miami"] = dataframe["Miami"].fillna(0).astype("int")
dataframe.head(3)
```

```python
dataframe["Rev Rank"] = dataframe["Miami"].rank(ascending = False).astype("int")
dataframe.head(10)
```

```python
dataframe.sort_values(by = "Miami", ascending = False)
```
